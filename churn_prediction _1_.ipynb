```python


import pandas as pd
import numpy as np
import random
churn_data = pd.read_csv("C:/Users/admin/Downloads/train.csv", header = 0)

X = churn_data.ix[:,0:-1]
Y = churn_data.ix[:,-1]

X.REPORTED_SATISFACTION = X.REPORTED_SATISFACTION.replace({'very_unsat':20, 
                                 'unsat':40, 
                                 'avg':60, 
                                 'sat':80, 
                                 'very_sat':100})


X.REPORTED_USAGE_LEVEL = X.REPORTED_USAGE_LEVEL.replace({'very_little':20, 
                                 'little':40, 
                                 'avg':60, 
                                 'high':80, 
                                 'very_high':100})


X.CONSIDERING_CHANGE_OF_PLAN = X.CONSIDERING_CHANGE_OF_PLAN.replace({'no':20, 
                                 'never_thought':40, 
                                 'considering':60, 
                                 'perhaps':80, 
                                 'actively_looking_into_it':100})

X = pd.get_dummies(X)

from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=19)

## Decision Tree
from sklearn import tree
class_weight = {0 : 40., 1: 43.}

dtree = tree.DecisionTreeClassifier(class_weight=class_weight, criterion='entropy', max_depth=4,
            max_features=None, max_leaf_nodes=None,
            min_impurity_decrease=1e-07, min_samples_leaf=1,
            min_samples_split=20, min_weight_fraction_leaf=0.0,
            presort=False, random_state=19, splitter='best')


dtree = dtree.fit(X_train, y_train)

dtree_predictions = dtree.predict(X_test)
print ("decision tree accuracy: ")
print (metrics.accuracy_score(y_test, dtree_predictions))
print (classification_report(y_test, dtree_predictions))


from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,dtree_predictions)
x=dtree.predict_proba(X_test)

churn_test_data = pd.read_csv("C:/Users/admin/Downloads/test.csv", header = 0)
churn_test_data.REPORTED_SATISFACTION = churn_test_data.REPORTED_SATISFACTION.replace({'very_unsat':random.randint(1,20), 
                                 'unsat':random.randint(20,40), 
                                 'avg':random.randint(40,60), 
                                 'sat':random.randint(60,80), 
                                 'very_sat':random.randint(80,100)})

churn_test_data.REPORTED_USAGE_LEVEL = churn_test_data.REPORTED_USAGE_LEVEL.replace({'very_little':random.randint(1,20), 
                                 'little':random.randint(20,40), 
                                 'avg':random.randint(40,60), 
                                 'high':random.randint(60,80), 
                                 'very_high':random.randint(80,100)})


churn_test_data.CONSIDERING_CHANGE_OF_PLAN = churn_test_data.CONSIDERING_CHANGE_OF_PLAN.replace({'no':random.randint(1,20), 
                                 'never_thought':random.randint(20,40), 
                                 'considering':random.randint(40,60), 
                                 'perhaps':random.randint(60,80), 
                                 'actively_looking_into_it':random.randint(80,100)})


churn_test_data = pd.get_dummies(churn_test_data)
final_pred = dtree.predict(churn_test_data)

final_pred_df = pd.DataFrame(data=final_pred,columns=["LEAVE"])

final_pred_df.to_csv('C:/Users/admin/Downloads/t_suchi.csv')
```
